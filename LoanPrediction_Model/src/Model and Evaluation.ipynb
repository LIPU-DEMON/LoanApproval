```python
import pandas as pd
from sklearn.model_selection import train_test_split,RandomizedSearchCV,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,make_scorer
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,classification_report
```

```python
data = pd.read_csv("Data/Transformation.csv")
data
```

```python
x = data.iloc[:,0:20]
Y = data.iloc[:,20]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size=0.2,random_state=23)
```

<h4>Logistic Regression</h4>

```python
LR_model = LogisticRegression()
LR = LR_model.fit(x_train,y_train)
LR
```

```python
Test = LR.predict(x_test)
Test
```

```python
accuracy = accuracy_score(y_test,Test)
precision_ = precision_score(y_test,Test,pos_label='Y')
print(accuracy)
print(precision_)
```

```python
#TUNING 
f1_ = make_scorer(f1_score, pos_label='Y')
params = {
    'C': [0.01, 0.1, 1, 10],
    "penalty":["l2","l1"],
    "class_weight":[None,"balanced"],
    "solver":["liblinear"],
    "max_iter":[100,200,300]
}
RSV = RandomizedSearchCV(LR_model,params,cv=5,scoring=f1_)
RSV = RSV.fit(x_train,y_train)
RSV
```

```python
RSV.best_params_
```

```python
LR_tune = LogisticRegression(solver='liblinear',penalty='l1',max_iter=100,class_weight=None,C= 1)
LR_tune = LR_tune.fit(x_train,y_train)
LR_tune
```

```python
Test_tune = LR_tune.predict(x_test)
Test_tune
```

```python
accuracy = accuracy_score(y_test,Test_tune)
precision_ = precision_score(y_test,Test_tune,pos_label='Y')
f1 = f1_score(y_test,Test_tune,pos_label="Y")
print(f1)
print(accuracy)
print(precision_)
```

## Decision Tree

```python
New_data = pd.read_csv("Data/transformation_org.csv")
New_data.head()
```

```python
X = New_data.iloc[:,0:20]
y = New_data.iloc[:,20]
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.20,random_state=33)
```

```python
dtc = DecisionTreeClassifier(criterion='entropy',min_samples_split=2,min_samples_leaf=8,max_depth=5)
dtc.fit(X_train,Y_train)
dtc
```

```python
predict = dtc.predict(X_test)
print("f1_score:",f1_score(Y_test,predict,pos_label="Y"))
print(accuracy_score(Y_test,predict))
print(precision_score(Y_test,predict,pos_label="Y"))
```

```python
#tuning
params = {
    "criterion":["gini","entropy","log_loss"],
    "splitter":["best","random"],
    "min_samples_split":[2,4,8,10,15],
    "min_samples_leaf":[2,5,8,10],
    "max_depth":[2,5,8,10]
}
f1_ = make_scorer(f1_score,pos_label="Y")
dtc_tune = RandomizedSearchCV(dtc,param_distributions=params,scoring=f1_,cv=5,n_iter=10)
dtc_tune.fit(X_train,Y_train)
dtc_tune
```

```python
best_model = dtc_tune.best_estimator_
predict = best_model.predict(X_test)
print("f1_score:",f1_score(Y_test,predict,pos_label="Y"))
print(accuracy_score(Y_test,predict))
print(precision_score(Y_test,predict,pos_label="Y"))
```

## Ensemble Technique: Stacking Classifier


```python
#splitting x_train and y_train 50/50
x_train_a,x_train_b,y_train_a,y_train_b = train_test_split(x_train,y_train,test_size=.50,random_state=23)
#base models
lg = LogisticRegression()
svc = SVC()
knn = KNeighborsClassifier()
#training base models
lg.fit(x_train_a,y_train_a)
svc.fit(x_train_a,y_train_a)
knn.fit(x_train_a,y_train_a)
#prediction on base models
pred_lg = lg.predict(x_train_b)
pred_svc = svc.predict(x_train_b)
pred_knn = knn.predict(x_train_b)
#stacking up those prediction 
input_features = np.column_stack((pred_lg,pred_svc,pred_knn))
#encoding input_features
encoders = []
input_encode = np.empty_like(input_features,dtype="int")
for i in range(input_features.shape[1]):
    lblencoder = LabelEncoder()
    input_encode[:,i] = lblencoder.fit_transform(input_features[:,i])
    encoders.append(lblencoder)
#train meta model dt using these input_features
dt = DecisionTreeClassifier()
dt.fit(input_encode,y_train_b)

# prediction on test
pred_lg_test = lg.predict(x_test)
pred_svc_test = lg.predict(x_test)
pred_knn_test = lg.predict(x_test)
#stack it up
input_features_test = np.column_stack((pred_lg_test,pred_svc_test,pred_knn_test))
#encoding 
encoders_test = []
input_encoder_test = np.empty_like(input_features_test,dtype="int")
for i in range(input_encoder_test.shape[1]):
    le = LabelEncoder()
    input_encoder_test[:,i] = le.fit_transform(input_features_test[:,i])
    encoders_test.append(le)
#final prediction
final_prediction = dt.predict(input_encoder_test)
final_prediction
```

```python
print(f1_score(y_test,final_prediction,pos_label="Y"))
print(accuracy_score(y_test,final_prediction))
print(precision_score(y_test,final_prediction,pos_label="Y"))
```
