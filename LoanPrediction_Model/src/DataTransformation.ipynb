```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as psx
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
data = pd.read_csv("Data/Validation.csv")
data.head()
```

```python
data.info()
```

```python
#Encoding Categorical Features using OnehotEncoder
categorical = data.iloc[:,[1,2,3,4,5,7]]
encode = OneHotEncoder(handle_unknown="ignore")
fit = encode.fit_transform(categorical)
encode_df = pd.DataFrame(fit.toarray(),columns=encode.get_feature_names_out(categorical.columns))
```

```python
encode_df
```

```python
new_data = pd.concat([encode_df.astype("int"),data.drop(columns=["Loan_ID","Gender","Married","Dependents","Education","Self_Employed","Property_Area"])],axis=1)
```

```python
new_data
```

```python
#Scaling
Std = StandardScaler()
Std_ = Std.fit_transform(new_data.loc[:,["ApplicantIncome","LoanAmount","Loan_Amount_Term","CoapplicantIncome"]])
df_std = pd.DataFrame(Std_,columns=Std.get_feature_names_out())
df_std
```

```python
New_data = pd.concat([new_data.drop(columns=["ApplicantIncome","LoanAmount","Loan_Amount_Term","CoapplicantIncome","Loan_Status"]),df_std,new_data["Loan_Status"]],axis=1)
```

```python
New_data
```

```python
New_data.to_csv("Data/Transformation.csv",index=False)
new_data.to_csv("Data/Transformation_org.csv",index=False)
```
