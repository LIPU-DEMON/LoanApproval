```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as psx
import plotly.graph_objects as go
```

```python
data = pd.read_csv("Data/Validation.csv")
data.head()
```

```python
data.isna().sum()
```

```python
data.drop(columns="Loan_ID",inplace=True)
```

<LI><B>RELATIONSHIPS BETWEEN DEPENDENT AND WITHIN INDEPENDENT FEATURES</B></LI>

```python
data.corr(numeric_only=True)
```

```python
data.head()
```

```python
Gender_married = data.groupby(["Gender","Married"]).size().reset_index(name="count")
Gender_married
```

```python
Status_GM = data.groupby(["Gender","Married","Loan_Status"]).size().reset_index(name="Count")
Status_GM
```

```python
Dependents_Status = data.groupby(["Dependents","Loan_Status"]).size().reset_index(name="Counts")
Dependents_Status
```

```python
Employed_Status = data.groupby(["Self_Employed","Loan_Status"]).size().reset_index(name="Counts")
Employed_Status
```

```python
applicant_avg = round(sum(data["ApplicantIncome"])/len(data["ApplicantIncome"]),2)
below_avg = data[data["ApplicantIncome"]<applicant_avg]
above_avg = data[data["ApplicantIncome"]>applicant_avg]
```

```python
b = below_avg.groupby(["ApplicantIncome","Loan_Status"]).size().reset_index(name="Counts")
below_status = b.groupby(["Loan_Status"])["Counts"].sum().reset_index(name="Total")
```

```python
#Loan_status on the basis of below avg applicant Income
below_status
```

```python
a = above_avg.groupby(["ApplicantIncome","Loan_Status"]).size().reset_index(name="Counts")
above_status = a.groupby(["Loan_Status"])["Counts"].sum().reset_index(name="Total")
```

```python
#Loan_status on the basis of above avg applicant Income
above_status
```

```python
Area_Status = data.groupby(["Property_Area","Loan_Status"]).size().reset_index(name="Counts")
Area_Status
```

```python
Loan_Term = data.groupby(["Loan_Amount_Term"])["LoanAmount"].mean().reset_index(name="Avg_Loan_Amt")
Loan_Term = pd.concat([Loan_Term.iloc[:,[0]],Loan_Term.iloc[:,[1]].astype("int")],axis=1)
Loan_Term
```

```python
lt = data.groupby(["Loan_Amount_Term"])["Loan_Status"].value_counts().unstack().reset_index()
lt = lt.iloc[:,[0,1,2]].replace(np.nan,0).astype("int")
lt
```

```python
merge = pd.merge(Loan_Term,lt,on="Loan_Amount_Term")
merge
```

```python
loan_status_count_AmtTerm = data.groupby(["Loan_Amount_Term"])["Loan_Status"].value_counts().reset_index()
```

<h1><b>UNIVARIATE ANALYSIS</b></h1>

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
data["Dependents"].value_counts().plot.pie(autopct="%1.1f%%",title="Dependents",ax=ax[0][0],explode=(0,0,0,0.1))
data["Self_Employed"].value_counts().plot.pie(autopct="%1.1f%%",title="Self_Employed",ax=ax[0][1],explode=(0,0.1))
data["Education"].value_counts().plot.pie(autopct="%1.1f%%",title="Education",ax=ax[1][0],explode=(0,0.1))
data["Property_Area"].value_counts().plot.pie(autopct="%1.1f%%",title="Property_Area",ax=ax[1][1])
plt.savefig("CategoricalUNIVARIATE.png",dpi=100)
plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,15))
plt.subplots_adjust(hspace=0.2, wspace=0.5)
data["ApplicantIncome"].plot.box(ax=ax[0][1],title="ApplicantIncome")
data["ApplicantIncome"].plot.hist(xlabel="applicant_income",bins=(50),ylabel="Counts",ax=ax[0][0],color="blue",title="ApplicantIncome")
data["LoanAmount"].plot.hist(xlabel="LoanAmount",bins=(50),ax=ax[1][0])
sns.violinplot(data["LoanAmount"],ax=ax[1][1])
ax[1,1].set_title("LoanAmount")
plt.savefig("NumericalUNIVARIATE.png",dpi=100)
#data["LoanAmount"].plot.violin(ax=ax[1][1],title="LoanAmount")
plt.show()
```

```python
fig,ax = plt.subplots(3,2,figsize=(15,18))
plt.subplots_adjust(hspace=0.3,wspace=0.5)
data.iloc[:,[0,1]].value_counts().plot.pie(autopct="%1.1f%%",title="MalesMarried/FemalesMarried",
                                          explode=(0.01,0.01,0.1,0.1),ax=ax[0,0])

sns.barplot(x=Status_GM["Loan_Status"],y=Status_GM["Count"],hue=Status_GM["Married"],ax=ax[0,1])
ax[0,1].set_title("LoanStatus")
sns.barplot(x=Dependents_Status["Loan_Status"],y=Dependents_Status["Counts"],hue=Dependents_Status["Dependents"],ax=ax[1,0])
ax[1,0].set_title("Loan_Status")
pie.plot.pie(subplots=True,autopct="%1.1f%%",ax=ax[1,1])
ax[1,1].set_title("Employed_Status")
sns.barplot(x=Area_Status["Loan_Status"],y=Area_Status["Counts"],hue=Area_Status["Property_Area"],ax=ax[2,0])
ax[2,0].set_title("Area_Status")
ax[2,1].axis("off")
plt.savefig("CategoricalBIVARIATE.png",dpi=100)
plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,15))
fig.subplots_adjust(hspace=0.3,wspace=0.5)
sns.barplot(x=below_status["Loan_Status"],y=below_status["Total"],ax=ax[0,0],palette=["green","lightgreen"])
ax[0,0].set_xlabel("LoanStatus < avg:5403.46")
ax[0,0].set_title("Count of ApplicantIncome_Status")
ax[0,0].set_ylim(0,438)
sns.barplot(x=above_status["Loan_Status"],y=above_status["Total"],ax=ax[0,1],palette=["lightblue","blue"])
ax[0,1].set_xlabel("LoanStatus > avg:5403.46")
ax[0,1].set_title("Count of ApplicantIncome_Status")
ax[0,1].set_ylim(0,438)
sns.lineplot(x=merge["Loan_Amount_Term"],y=merge["Avg_Loan_Amt"],ax=ax[1,0])
ax[1,0].set_title("AvgLoanAmt vs LoanStatus")
sns.lineplot(x=loan_status_count_AmtTerm["Loan_Amount_Term"],y=loan_status_count_AmtTerm["count"],style=loan_status_count_AmtTerm["Loan_Status"],size=loan_status_count_AmtTerm["Loan_Status"],ax=ax[1,1])
ax[1,1].set_title("LoanStatus vs LoanAmtTerm")
plt.savefig("ContinuosBIVARIATE.png",dpi=100)
plt.show()
```
